# 動的計画法と最適化

## 動的計画法

直感的な理解としては、全体が最適ならばその部分部分も最適な状態を満たしている、と推測できる.


### 後ろ向き帰納法

時間軸の最後からその時間における最適解を関数として求め、それをひとつ前の時間軸に代入して関数解を求めることを繰り返すことで最適解を求める手法. 最適解は初期状態に応じて決まる.

例として、ある個人Aさんがある財Xを一定期間で消費する簡単なモデルを考える.

t期のAさんの効用はt期の財Xの消費量$Q_t=X \cdot q_t$,$(0\leq q_t \leq 1-\sum_{i=1}^{i=t-1}q_i)$,ただし$t=1$で$0 \leq q_t \leq 1)$によって決まる$U_t(Q)=\sqrt{Q}$という関数で表現できる.
Aさんが3期である財を消費しきるときAさんはどのように各期の消費量$q_t$を決めるか.

t期にAさんが直面する財Xの残量を$S_t$と置くと、$S_1=X,S_4=0$である.

Aさんは効用を最大化させるため次のような最適化問題を解く.

$$
max \sum_{i=1}^3 \sqrt{U_i(q_i)}\\
\begin{aligned}
s.t S_{t+1}& = S_t - Xq_t \\
S_0&=X \\
S_4&=0
\end{aligned}
$$

以下ではt期の価値関数を$V_T$であらわす.

$\
V_3 = max\{ \sqrt{U_3(Xq_3)} + V_4\} = max\{ \sqrt{U_3(Xq_3)}\} (\because V_4=0, since S_4=0) \\
V_2 = max\{ \sqrt{U_2(Xq_2)} + V_3\} = max\{ \sqrt{U_2(Xq_2)} + \sqrt{U_3(Xq_3)}\}$

3期には2期の残量すべてを消費するはずなので$q_3=S_3=S_2-Xq_2$となり、$V_2=max \{ \sqrt{X \cdot q_2}+\sqrt{S_2-Xq_2}\}$

これを最大にするような$q_2$を求めればいいので、$\frac{\partial V_2}{\partial q_2}=\frac{1}{2}X(Xq_2)^{\frac{-1}{2}}-\frac{1}{2}X(S_2-Xq_2)^{\frac{-1}{2}}=0$

すなわち$q_2=\frac{S_2}{2X}$

同様に$V_1$を考えると、$V_1=max \{ \sqrt{Xq_1}+V_2 \}=max \{ \sqrt{Xq_1} + \sqrt{Xq_2}+\sqrt{S_2-Xq_2}\}\\
= max \{ \sqrt{Xq_1}+2\sqrt{\frac{S_2}{2}}\}=max \{ \sqrt{Xq_1}+2\sqrt{\frac{S_1-Xq_1}{2}} \}$

$q_1$について偏微分してこれを最大化する$q_1$を求めると、$\frac{\partial V_1}{\partial q_1}=\frac{1}{2}X(Xq_1)^{\frac{-1}{2}}-\frac{1}{2}X(\frac{S_1-Xq_1}{2})^{\frac{-1}{2}}=0$

すなわち$3Xq_1=S_1$,また、$S_1=X$であるから$q_1=\frac{1}{3}$となる. よって$q_1=q_2=q_3=\frac{1}{3}$となる.

つまり、各期間で財の1/3ずつ消費することで効用を最大化できることがわかる.

### 動的計画法のプログラムの実装

動的計画法を解くことと漸化式を解くことはほぼほぼ等しい. 

たとえば、よくある例題だがフィボナッチ数列$fib_0=1,fib_1=1,f_n=fib(n-1)+fib(n-2)$は素直に書くと以下のように書ける.

```python
cnt : int = 0

    
def fib(n : int)-> int:
    global cnt
    cnt += 1
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fib(n-1)+fib(n-2)
    
    
print(fib(6))
print(cnt)
```

この書き方のメリットは、漸化式を素直な形で書けるという点である.
しかしこの書き方は`fib(n)`を呼び出すまでに`fib(k)`(k<n)を複数回計算しているという点で効率が悪い. `cnt`という変数で`fib`が呼び出されている回数をカウントしてみると,`fib(6)`を得るために25回も`fib`が呼ばれている.

そこでこれを後ろ向き帰納法のアイディアを使って書いてみよう.(本質的にはメモ化再帰と同じ)

`fib(2)`について初期値が与えられているのであとは得られた値を使って順に代入していけばいい.

一方でこちらの解法は長さ`n`の配列が必要になる. 時間計算量と空間計算量にはトレードオフの関係があることを覚えておくといいだろう.

動的計画法というと再帰のイメージが強いがそれまでに得られた答えをうまく再利用して次の解を得ると考えておけば再帰を使わなくても定式化できる.

得られたところから配列の値を埋めていく感覚を覚えておこう.

```python
def fib(n:int)-> int:
    global cnt
    fibs:list[int]= []
    fibs+=[0]
    fibs+=[1]
    for idx in range(2,n+1):
        print(fibs)
        fibs+=[fibs[idx-1] + fibs[idx-2]]
    return fibs[n]
print(fib(6))   
```

### 直線探索法(linear search)

> 最急降下法や準ニュートン法などの最適化手法において得られた解の探索方向に対して、適切な更新幅を求める手法である.  [直線探索を使った最急降下法をPythonで実装](https://helve-blog.com/posts/math/gradient-descent-armijo/) より 2021/05/20 閲覧

勾配を移動するステップ幅が大きすぎると解が定まらず、小さすぎると解に収束しない、または収束が無駄に遅くなると直感できる。

このステップ幅は深層学習の分野では学習率とも呼ばれる. 巷にある深層学習のコードを眺めてみるとステップ幅が0.1や0.01で固定されている場合もしばしばある. 計算資源が潤沢にあるならば複雑なアルゴリズムを書くよりもその資源を活用するほうが都合がいいかもしれない.

### ニュートン法・準ニュートン法

ニュートン法では点$x_k$における目的関数$f(x)$の二次多項式近似の勾配を$f(x_k)$の2次までのテイラー展開から求め、ステップ幅を$\alpha$として勾配がもっとも大きくなる方向に$\delta\cdot\alpha$だけ$x$を動かす処理を,$x$と更新後の$x'=(x+\alpha\delta)$が十分近くなる、つまり$f(x)$と$f(x')$の差が十分小さくなるまで繰り返す. (勾配ベクトルが0になる場合を考えても良い.) 

直線探索法では$\delta=\alpha f'(x)$だが、ニュートン法では$x_k$で求めたヘッセ行列を$H(x_k)$として$\delta = \alpha H(x_k)^{-1}f'(x)$とする.







#### アルミホ条件(armijo condition/rule)とウルフ条件(wolfe condition/rule)
上で述べたステップ幅に関する制約を数学的に記述したものがアルミホ条件とウルフ条件である.

アルミホ条件とは点$x$での勾配ベクトルを$\nabla f(x)$,定数$\xi$を$0<\xi<1$として以下のような式を満たすステップ幅$\alpha$を選ぶものである.

$$
f(x_k + \delta )\leq f(x_k) + \xi_1\alpha \nabla f(x_k)^T\delta
$$


ウルフ条件はアルミホ条件に加えて、 ただし$0<\xi_1 < \xi_2 < 1$をみたす定数$\xi_2$のもとで$f'(x+\alpha\delta)^T \delta \leq \xi_2 f'(x)^T\delta$を満たす制約を$\alpha$に与える. これによって、アルミホ条件だけの場合に$\alpha$がかぎりなく小さくなることを避けることができる.

### ニュートン法の実装

$x=\begin{pmatrix}x_1 \\ x_2 \end{pmatrix}$について$\mathbb{R}^2 \rightarrow R$の関数$f(x)=3x_1^2 + 2x_1x_2 + x_2^2$をニュートン法で最小化する.

行列をPythonの組み込み配列であらわす場合は`list[list[float|int]]`であらわす. 内側の`list[float|int]`が行の成分である. つまり列ベクトルは`[[x_1],[x_2]]`とあらわす.

今回は微分とヘッセ行列を手計算で求めたが、十分小さな実数$h$(例えば、`1e-10`)について$\frac{f(x+h)-f(h)}{h}$を計算しても良い. (ただし、小さすぎる場合数値計算の誤差でかえって悪い結果になる場合がある.)
また高次の関数をあつかう場合、ヘッセ行列を$f(x_k)$のヘッセ行列を$f(x_k)$でのテイラー展開から求める.


なおニュートン法では$x_k$についてのヘッセ行列$\nabla^2f(x_k)$と`g(x)`の連立方程式を解くので、ヘッセ行列の逆行列の計算や行列の保持で計算量やメモリが必要である. 準ニュートン法では、ヘッセ行列を近似することでこのデメリットを抑えている.

```python
from typing import Callable
from __future__ import annotations
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

まず、$f(x)$,$f(x)$の勾配ベクトル$\nabla f(x)=g(x)$,ヘッセ行列$\nabla^2f(x)=hessian$を定義する.

```python
# f(x):R^2 -> R 
f : Callable[[list[list[float|int]]],float|int]=(
    lambda x : 3* pow(x[0][0],2) + 2*x[0][0]*x[1][0] + pow(x[1][0],2))
# g(x): R^2 -> R^2
g :Callable[[list[list[float|int]]],list[list[float|int]]] =(
    lambda x :np.array([[ 6*x[0][0]+2*x[1][0]], [2*x[0][0] + 2*x[1][0]]])  ) 
# 2x2 matrix
hessian:list[list[float|int]]=np.array([[6,2],[2,2]])
```

$f(x)$は次のような形である. 上の図が$f(x)$をプロットしたもの、下の図が$f(x)$の等高線を二次元にプロットしたものである. (0,0)付近で最小となることがみてとれる.

```python

x1 = np.linspace(-5, 5, 30)
x2 = np.linspace(-5, 5, 30)
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111,projection='3d')
ax.view_init(elev=15, azim=-100)

x1_mesh, x2_mesh = np.meshgrid(x1, x2)
z =f(np.array(([x1_mesh],[x2_mesh])))
ax.plot_surface(x1, x2, z, cmap = "coolwarm")
ax.contour(x1, x2, z, colors = "black", offset = -1)
```

```python

z =f(np.array(([x1_mesh],[x2_mesh])))

fig, ax = plt.subplots(figsize=(6, 6))
ax.contour(x1, x2, z, levels=np.logspace(-0.3, 1.2, 10))
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_aspect('equal')
plt.show()
```

```python
def newtonMethod(
    f:Callable[[list[list[float|int]]],float|int],# f(x):R^2->R
    x: list[list[float|int]],# x_0:R^2,
    g: list[list[float|int]],# ▽f(x)
    hessian: list[list[float|int]]): # ▽^2f(x)
    threshold:float = 1e-12
    f_prev :float= 1e12
    diff:float = 1e12
    xi_1=0.3
    xi_2 = 0.4
    tau = 0.99
    while diff > threshold:
        alpha:float = 1.0
        f_x = f(x)
        # caluculate delta
        delta = -np.dot(np.linalg.inv(hessian),g(x))
        # update alpha in accordance with armijo rule and wolfe rule
        # while 文を使っているのでアルミホ条件だけでもアルミホ条件を最初に満たしたalphaが選択されるので極端に小さいalphaが選ばれるのは避けられる.
        # ただしその場合ウルフ条件を満たす可能性は高いが必ず満たすとは言えない.
        while f(np.array(x) + alpha*delta) > (f(x) + xi_1*alpha*np.dot(g(x).T,delta)) and np.dot(g(x+alpha*delta).T,delta) >=xi_2*np.dot(g(x).T,delta):
            alpha = tau*alpha
        x += alpha*delta
        x= x.tolist()
        diff = abs(f_x - f_prev)
        f_prev = f_x
    return x,f(x)
newtonMethod(f,[[10],[-13]],g,hessian)
```

真の解が$(0,0)$で$0$,得られた解が$(0.0,-8.758\cdot10^{-47})$で$7.670\cdot10^{-93}$なのでうまく近似的な解を求められた.
