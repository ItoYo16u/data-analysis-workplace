## 解答

### 1回
(a)
$$
P(X=0) = \sum_{y=0}^2 P(X=0,Y=y) = \frac{1}{8} + \frac{1}{8} + \frac{1}{4} = \frac{1}{2}
$$

$$
P(X=1) = \sum_{y=0}^2 P(X=1,Y=y) = \frac{1}{8} + \frac{1}{8} + \frac{1}{4} = \frac{1}{2}
$$

(b) X=0を与えた時の Y の条件付確率関数は次のようになる.

$$
P(Y=0|X=0)= \frac{P(X=0,Y=0)}{P(X=0)} = \frac{1/8}{1/2} = \frac{1}{4}
$$

$$
P(Y=1|X=0)= \frac{P(X=0,Y=1)}{P(X=0)} = \frac{1/8}{1/2} = \frac{1}{4}
$$

$$
P(Y=2|X=0)= \frac{P(X=0,Y=2)}{P(X=0)} = \frac{1/4}{1/2} = \frac{1}{2}
$$

同様に X=1 を与えた時の条件付確率関数は以下の通り.

$$
P(Y=0|X=1)=\frac{1}{4}, P(Y=1|X=1)=\frac{1}{4},P(Y=2|X=1)=\frac{1}{2}
$$

(c)
定義より.

$$
E(Y) = \sum_{y=0}^2 y P(Y=y) = 0 \cdot \frac{1}{4} + 1 \cdot \frac{1}{4} + 2 \cdot \frac{1}{2} = \frac{5}{4}
$$

(d)

定義より.

$$
E(Y|X=0) = \sum_{y=0}^2 y P(Y=y|X=0) \\
  = 0 \cdot P(Y=0|X=0) + 1 \cdot P(Y=1|X=0) + 2 \cdot P(Y=2 |X=0) \\
  = 0 \cdot \frac{1}{4} + 1 \cdot \frac{1}{4} + 2 \cdot \frac{1}{2} = \frac{5}{4}
$$

同様に

$$
E(Y|X=1) = \sum_{y=0}^2 y P(Y=y|X=1) \\
  = 0 \cdot P(Y=0|X=1) + 1 \cdot P(Y=1|X=1) + 2 \cdot P(Y=2 |X=1) = \frac{5}{4}
$$

従って

$$
E(Y|X)=
    \begin{cases}
        \frac{5}{4}  &   \text{if X=0}  \\
        \frac{5}{4}        &   \text{if X=1}
    \end{cases}
$$

つまり、 $E(Y|X)= 5/4$ .

### 2回

(a) まず $P(X=0) = \sum_y P(X=0,Y=y) = P(X=0,Y=-1) + P(X=0,Y=1) = 2/3$ に注意する.   条件付確率は以下の通り.

$$
P(Y=-1,X=0) = \frac{P(X=0,Y=-1}{P(X=0)} = \frac{ 1/3}{ 2/3} = \frac{ 1 }{ 2 }
$$
$$
P(Y=0,X=0) = \frac{P(X=0,y-1)}{P(X=0)} = \frac{0}{2/3}= 0 \\
$$
$$
P(Y=1|X=0) = \frac{P)(X=0,Y=1)}{P(X=0)} = \frac{1/3}{2/3} = \frac{1}{2}
$$

よって、条件付確率関数 $P(Y=y,X=0)$ は次のように書ける.

$$
\label{eq: cases P}
P(Y=y|X=0)=
    \begin{cases}
        1  &   \text{if y = -1,1}  \\
        0        &   \text{otherwise}
    \end{cases}
$$

(b) (a) と同様に $P(X=1) = \sum_y P(x-1,Y=y) = P(X=1,Y=0) = 1/3$ に注意すれば、次の結果を示せる.

$$
\label{eq: cases f}
P(Y=y|X=1)=
    \begin{cases}
        1  &   \text{if y = 0}  \\
        0        &   \text{otherwise}
    \end{cases}
$$

(c) たとえば、 (a) で示したように $P(Y=0,X=0)=0$ だが $P(Y=0) = P(X=1,Y=0) = 1/3$ であるため、$P(Y=0 |X=0) \neq P(Y=0)$ であり、これは X と Y が独立でないことを意味する.

(d) Y が -1,0,1 の3つの実現値をとる確率変数であることに注意しつつ、条件付確率の定義通り計算する.

$$
E(Y|X=0) = \sum_{y=-1}^{1} y \cdot P(Y=y|X=0) = -1 \cdot \frac{1}{2} + 0 \cdot 0 + 1 \cdot \frac{1}{2} = 0
$$

$$
E(Y|X=1) = \sum_{y=-1}^{1} y \cdot P(Y=y|X=1) = -1 \cdot 0 + 0 \cdot 1 + 1 \cdot 0 = 0
$$

その結果、 X を条件付けあ¥他時の Yの条件付き期待値は次のとおりである.


$$
E(Y|X)=
    \begin{cases}
        0  &   \text{if X = 0}  \\
        0        &   \text{otherwise (X=1)}
    \end{cases}
$$

つまり,$E(Y|X) = 0$. E(Y) は繰り返し期待値の法則から求められる.


$$
E(Y) = E(E(Y|X)) = \sum_{x_0}^1 E(Y|X=x)P(X=x) = E(Y|X=0)P(X=0) + E(Y|X=1)P(X=1)=0
$$ 

解説: この XとYは独立ではないが無相関な確率変数の一例となっている. 
実際に繰り替えし期待値の法則と関係 $E(Y|X)=E(Y)$ を使えば、$Cov(X,Y) = E(XY) - E(X)E(Y)  = E(E(XY|X)) - E(X)E(Y) = E(X\cdot E(Y|X)) - E(X)E(Y) = E(XE(Y)) - E(X)E(Y) = E(X)E(Y) - E(X)E(Y) = 0$


### 第3回

1

まず、任意の実数 $a,b \neq 0$ について、期待値の線形性より、

$$
E(Z) = \frac{E(X) + a}{b} = \frac{\mu + a}{b}
$$

である.
よって、$\alpha = -\mu$とすれば、 $b \neq 0$ の値によらず $E(Z)=0$となる.
次に分散の性質より,

$$
V(Z) = \frac{V(X)}{b^2} = \frac{\sigma^2}{b^2}
$$

である. よって $b = \sigma$ とすれば、 $V(Z)=1$ となる. つまり、もとめる確率変数は $(X-\mu)/\sigma$

2
$N(0,1)$ の上側 $\alpha$点の定義から明らかに

$$
P(Z>1.96) = 0.025
$$

標準正規分布$N(0,1)$の対称性から
$$
P(Z < -1.96) = 0.025
$$

である.

事象A についての確率の性質、 $P(A) = 1 - P(A^c)$から

$$
P(-1.96 \leq Z \leq 1.96) = 1 - P(Z>1.96 or Z < -1.96) = 1 - P[P(Z>1.96)+P(Z < -1.96)] = 1 -0.05 = 0.95

### 第4回


(a) 標準誤差 $SE(\bar{x}) =  \sqrt{ S_x^2/n}$ より、標準誤差の値は $se(\bar{x}) = \sqrt{s_x^2/n}  =  \sqrt{0.0025} = 0.05$ と求められる.  

母平均 $\mu x$ の $1-\alpha$ 信頼区間は、以下のように与えられる.

$$
  [ \bar{X} - z_{\alpha/2} SE(\bar{X}), \bar{X} + z_{\alpha/2} SE(\bar{X})]
$$

$\alpha = 0.05$ と 各推定値を当てはめれば、母平均 $\mu x$ の 95% 信頼区間の値は以下の通り.

$$
[ 0.5 - 1.96 \cdot 0.05, 0.5 + 1.96\cdot0.05] = [0.402,0.598]
$$

(b) t 統計量は以下を満たす.

$$
T = \frac{\bar{X}- \mu x}{SE(\bar{X})} \xrightarrow{d} N(0,1)
$$

$H_0$ のもとでの t 統計量の値は、
$$
t = \frac{\bar{x} - 0}{se(\bar{X})} = \frac{0.5 - 0}{0.05} = 10
$$

この t 値(の絶対値)は N(0,1) の上側 2.5% 点 $z_{0.025} = 1.96$ よりおおきいため、 優位水準5%のもとで $H_0$は棄却される. 

### 第6回

1

$H_0: \mu x = 0.5, H_1: \mu x \neq 0.5$

a. 標本分散については Xが0 または 1しかとらないことから $S_X^2=n^{-1}\sum_{i=1}^n X_i^2 - (\bar{X})^2=n^{-1}\sum_{i=1}^n X_i - (\bar{X})^2=\bar{X}-(\bar{X})^2$ となることに注意.

$$
\bar{x} = \frac{1}{n} \sum{i=1}^n x_i = \frac{50}{100} = 0.5
$$

$$
s_x^2 = \bar{x} - (\bar{x})^2 = 0.5 -0.25 = 0.25
$$

$$
se(\bar{x}) = \sqrt{\frac{s_x^2}{n}} = \frac{0.5}{10} = 0.05
$$

$\mu x$ の95%信頼区間の推定値は以下の通り.

$$
t = \frac{\bar{x} - \mu x}{se(\bar{x}} = \frac{0.5-0.5}{0.05} = 0
$$

$t > z_{0.025}$ または $t < -z_{0.025}$ ではないため, $H_0$ は有意水準5%で棄却されない

p値は以下の通り

$$
p = 2 \cdot \phi (-|t|) = 2 \cdot \phi(0) = 2 \cdot 0.5 = 1
$$

$\phi (0) = 0.5$は標準正規分布N(0,1)の密度関数が0を中心として対象であることを利用している.

### 第7回

$$
\sum_{i=1}^n(Y_i - \hat{\beta}X_i = 0
$$

これを$\hat{\beta}$について解けば、最小二乗推定量$\hat{\beta}$の式は以下のようになる.

$$

$$

を得る.


### 第8回
a. 省略

b.

最小二乗推定量の式に回帰モデルを代入すれば

$$
\hat{\beta} = \frac{\sum_{i=1}^n X_i(\beta X_i + U_i} {\sum_{i=1}^n X_i^2} = \beta + \frac{\sum_{i=1}^n X_i U_i}{X_i^2}
$$

説明変数 $X_1,\dots,X_n$ を与えた時の条件付き期待値を求めると

$$
E(\hat{\beta}|X_1,\dots,X_n) = \beta + \frac{\sum{i=1}^n X_i E(U_i|X_1,\dots,X_n)}{sum_{i=1}^n X_i^2}   = \beta + \frac{\sum_{i=1}^n X_i E(U_i|X_i)}{\sum_{i=1}^n X_i^2} = \beta
$$

であるため、繰り返し期待値の法則から

$$
E(\hat{\beta}) = E[E(\hat{\beta}|X_1,\dots,X_n)] = E[\beta] = \beta
$$

これは最小二乗推定量 $\hat{\beta}$ が回帰係数の不偏推定量であることを意味する.

c.

大数の法則より以下が成り立つ.

$$
\frac{1}{n} \sum_{i=1}^n X_i U_i \xrightarrow{p} E(X_i U_i), \frac{1}{n} \sum_{i=1}{n} X_i^2 \xrightarrow{p} E(X_i^2)
$$

が成り立つ. ここで外生性 $E(U_i|X_i)=0$のもとで繰り返し期待値の法則から $E[X_i U_i]= E[X_i E[U_i|X_i]] = 0$  が成り立つ.

これらの結果と連続写像定理から

$$
\hat{\beta} = \beta + \frac{n^{-1}\sum_{i=1}^n X_i U_i}{n^{-1}\sum_{i=1}^nX_i^2} \xrightarrow{p} \beta + \frac{E(X_i U_i)}{E(X_i^2)} = \beta
$$

が成り立つ. これは最小二乗推定量$\hat{\beta}$ が回帰係数 $\beta$の一致推定量であることを意味する.

### 第11回

### 第12回

a. 繰り返し期待値の法則から以下を得る.

$$
\begin{split}
E(RZ) &= E(E(RZ|Z)) = \sum_{i=0}^1 E(RZ|Z=z)P(Z=z)\\
&= E(RZ|Z=0)P(Z=0) + E(RZ|Z=1)P(Z=1) = E(R|Z=1)P(Z=1)
\end{split}
$$

$$
E(R) = E(E(R|Z)) = \sum_{z=0}^1 E(R|Z=z)P(Z=z) \\
= E(R|Z=0)P(Z=0)+ E(R|Z=1)P(Z=1)
$$

また、 Zが0か1をとることと期待値の定義より

$$
E(Z) = \sum_{z=0}^1 zP(Z=z) \\
= 0 \cdot P(Z=0) + 1 \cdot P(Z=1)\\
= P(Z=1)
$$

共分散の式に以上の関係を代入すると以下が得られる.

※ $P(Z=0)+ P(Z=1)= 1$ に注意せよ.

$$
\begin{split}
Cov(R,Z) &= E(RZ) - E(R)E(Z)\\
&= E(R|Z=1)P(Z=1) - \{E(R|Z=0)P(Z=0)+ E(R|Z=1)P(Z=1)\}\cdot P(Z=1)\\
&= P(Z=1)\{E(R|Z=1)- E(R|Z=1)P(Z=1) - E(R|Z=0)P(Z=0) \}\\
&= P(Z=1)\{E(R|Z=1) - E(R|Z=1)(1-P(Z=0)) - E(R|Z=0)P(Z=0)\}\\
&= P(Z=1)P(Z=0)\{E(R|Z=1)- E(R|Z=0)\}
\end{split}
$$

b. (a)の結果に R=Y または R= Xを代入すると明らか.

$$

$$

$$
Cov(X,Z)= P(Z=0)P(Z=1) \{E(X|Z=1)-E(X|Z=0)\}
$$
