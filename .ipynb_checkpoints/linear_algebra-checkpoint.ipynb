1. 解

$A^T=A$よりAは実対称行列であるから直交行列を用いて対角化できる.

固有方程式$|A-\lambda I|=0$を解くと、$\lambda=1,6$. 固有方程式が異なる2つの実数解をもつので、固有ベクトル$\begin{bmatrix}1 \\ -2 \end{bmatrix}$,$\begin{bmatrix}2 \\ 1 \end{bmatrix}$を大きさ1になるように正規化したベクトルを$p_1,p_2$と置くと、$p_1\cdot p_2=0$となる. すなわち$p_1,p_2$は正規直交基底となるので直交行列$P=(p_1,p_2)=\frac{1}{\sqrt{5}}\begin{bmatrix} 1 & 2 \\ -2 &1 \end{bmatrix}$を用いて,$P^{-1}AP=P^TAP=\begin{bmatrix} 1 & 0 \\ 0 & 6 \end{bmatrix}$と対角化できる.


note: n次対称行列が相異なる実数固有値をn個もつとき、大きさ1に正規化した固有ベクトルから成る直交行列$L$の逆行列は$L^T$となる. 固有値が重解を持つ場合は、対角行列に近い形のジョルダン標準形を求める.

2. 解

$A^T=A$よりAは実対象行列であるから直交行列を用いて対角化できる.

固有方程式$|A-\lambda I|=(1-\lambda)(2-\lambda)(3-\lambda)=0$より$\lambda=1,2,3$. 正規化した固有ベクトルは$\frac{1}{\sqrt{2}}\begin{bmatrix} 1 \\ -1 \\ 0 \end{bmatrix},\frac{1}{\sqrt{2}}\begin{bmatrix} 0 \\ 0 \\ \sqrt{2} \end{bmatrix},\frac{1}{\sqrt{2}}\begin{bmatrix} 1 \\ 1 \\ 0 \end{bmatrix}$.

行列Aは直交行列$P=\frac{1}{\sqrt{2}}\begin{bmatrix}1 & 0 & 1 \\ -1 & 0 & 1 \\ 0 & \sqrt{2}  & 0 \end{bmatrix}$をもちいて$P^{-1}AP=P^TAP=\begin{bmatrix}1 & 0 & 0 \\ 0 & 2 & 0 \\ 0 & 0 & 3 \end{bmatrix}$に対角化できる.



### 二次形式
与えられた多項式は以下のように二次形式であらわせる.

1. $\begin{bmatrix}x & y\end{bmatrix}\begin{bmatrix}1 & -1 \\ -1 & 0 \end{bmatrix}\begin{bmatrix} x \\ y \end{bmatrix}$
2. $\begin{bmatrix}x & y & z\end{bmatrix}\begin{bmatrix}1 & -1 & 0 \\ -1 & 0 & -1 \\ 0 & -1 & 1 \end{bmatrix}\begin{bmatrix} x \\ y \\ z \end{bmatrix}$

与えられた行列は任意の$b_1,b_2,b_3\neq0$について以下の符号が成立する.

1. $\begin{bmatrix} b_1 & b_2 \end{bmatrix} \begin{bmatrix} 1 & 1 \\ 1 & 1 \end{bmatrix} \begin{bmatrix} b_1 \\ b_2 \end{bmatrix}=(b_1+b_2)^2 \geq0$. 任意のbについて$b'Ab$が成立するので半正定値.
2. $\begin{bmatrix} b_1 & b_2 & b_3 \end{bmatrix} \begin{bmatrix} 2 & 1 & 0 \\ 1  & 2 & 1 \\  0 & 1 & 2 \end{bmatrix} \begin{bmatrix} b_1 \\ b_2 \\ b_3 \end{bmatrix}=(b_1+b_2)^2 + (b_2+b_3)^2 + b_1^2 + b_3^2 \gt0$. $b\neq9$である限り必ず$b'Ab>0$が成立するので正定値.
